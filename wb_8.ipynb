# Task 1.1.1

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans

X = np.array([[5,3],
[10,15],
[15,12],
[24,10],
[30,45],
[85,70],
[71,80],
[60,78],
[55,52],
[80,91],])

kmeans = KMeans(n_clusters=5, random_state=0)
clusters = kmeans.fit_predict(X)
kmeans.cluster_centers_.shape

plt.scatter(X[:, 0], X[:, 1], c = clusters, s = 20, cmap='summer')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'blue', s = 100, alpha = 0.9)
plt.show()
```

# Task 1.1.1(1)

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans

from sklearn.datasets import load_iris
iris = load_iris()
iris.data.shape

kmeans = KMeans(n_clusters=5, random_state=0)
clusters = kmeans.fit_predict(iris.data)
kmeans.cluster_centers_.shape

plt.scatter(iris.data[:, 0], iris.data[:, 1], c = clusters, s = 20, cmap='summer')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'blue', s = 100, alpha = 0.9)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

#зарузим набор данных
import seaborn as sns
customer_data = sns.load_dataset('iris')
customer_data.head()


data = customer_data.iloc[:, 2:4].values

import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title('Customer Dendgrams')
dend = shc.dendrogram(shc.linkage(data, method='ward'))


from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)

plt.figure(figsize=(10, 7))
plt.scatter(data[:, 0], data[:, 1], c=cluster.labels_, cmap='rainbow')
plt.show()
```
